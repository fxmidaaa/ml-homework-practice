```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = 'credit_card_default.xls'
```

```python
df = pd.read_excel(data)
```

```python
df.head()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')

new_columns = list(df.iloc[0].values)
new_columns[0] = 'id'
df.columns = new_columns
df = df.drop(df.index[0]).reset_index(drop=True)
```

```python
df.head().T
```

```python
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df = df.drop(columns=['ID'])
```

```python
df.isna().sum()
```

```python
df
```

```python
list(df.columns)
```

```python
df.nunique()
```

```python
categorical = ['sex', 'education', 'marriage', 'pay_0', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']
numerical = ['limit_bal', 'age', 'bill_amt1', 'bill_amt2', 'bill_amt3', 'bill_amt4', 'bill_amt5', 'bill_amt6',
             'pay_amt1', 'pay_amt2', 'pay_amt3', 'pay_amt4', 'pay_amt5', 'pay_amt6']

# Our target variable is default_payment_next_month 
```

```python
corr_matrix = df[numerical].corr()
corr_matrix
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state = 42)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state = 42)

len(df_train), len(df_val), len(df_test)
```

```python
df_train = df_train.reset_index(drop=True)
df_val =df_val.reset_index(drop=True)
df_test= df_test.reset_index(drop=True)

y_train = df_train.default_payment_next_month
y_val = df_val.default_payment_next_month
y_test = df_test.default_payment_next_month

del df_train['default_payment_next_month']
del df_val['default_payment_next_month']
del df_test['default_payment_next_month']
```

```python
from sklearn.metrics import mutual_info_score

def mut_info_target(series):
    return mutual_info_score(series, y_train)

mi = df_train[categorical].apply(mut_info_target)

mi.sort_values(ascending=False)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction import DictVectorizer

model = LogisticRegression(solver='liblinear', random_state = 42, max_iter=30000)
dv = DictVectorizer(sparse=False)

train_dicts = df_train[categorical + numerical].to_dict(orient='records')
val_dicts = df_val[categorical + numerical].to_dict(orient='records')
```

```python
X_train = dv.fit_transform(train_dicts)
X_val = dv.transform(val_dicts)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
default_decision = (y_pred >= 0.5)

original_accuracy = (y_val == default_decision).mean()

original_accuracy
```

now, let's get rid of most useless features and check our accuracy.
in mutual_info_score, we see that marriage, sex - variables with the lowest score.

```python
train_dicts = df_train[categorical + numerical].drop(columns=['marriage']).to_dict(orient='records')
val_dicts = df_val[categorical + numerical].drop(columns=['marriage']).to_dict(orient='records')

X_train = dv.fit_transform(train_dicts)
X_val = dv.transform(val_dicts)

model.fit(X_train, y_train)
y_pred = model.predict_proba(X_val)[:, 1]

new_decision = (y_pred >= 0.5)
new_accuracy = (y_val == new_decision).mean()

abs(original_accuracy - new_accuracy)
```

```python
train_dicts = df_train[categorical + numerical].drop(columns=['sex']).to_dict(orient='records')
val_dicts = df_val[categorical + numerical].drop(columns=['sex']).to_dict(orient='records')

X_train = dv.fit_transform(train_dicts)
X_val = dv.transform(val_dicts)

model.fit(X_train, y_train)
y_pred = model.predict_proba(X_val)[:, 1]

new_decision = (y_pred >= 0.5)
new_accuracy = (y_val == new_decision).mean()

abs(original_accuracy - new_accuracy)
```

```python
train_dicts = df_train[categorical + numerical].drop(columns=['education']).to_dict(orient='records')
val_dicts = df_val[categorical + numerical].drop(columns=['education']).to_dict(orient='records')

X_train = dv.fit_transform(train_dicts)
X_val = dv.transform(val_dicts)

model.fit(X_train, y_train)
y_pred = model.predict_proba(X_val)[:, 1]

new_decision = (y_pred >= 0.5)
new_accuracy = (y_val == new_decision).mean()

abs(original_accuracy - new_accuracy)
```

since these features have such a low prediction score, they don't impact on the decision since it doesn't cross the threshold of 0.5

```python
train_dicts = df_train[categorical + numerical].to_dict(orient='records')
val_dicts = df_val[categorical + numerical].to_dict(orient='records')
X_train = dv.fit_transform(train_dicts)
X_val = dv.transform(val_dicts)

for regularization in [0.01, 0.1, 1, 10, 100]:
    model = LogisticRegression(solver='liblinear', C=regularization, max_iter = 1000, random_state = 42)
    model.fit(X_train, y_train)
    y_pred = model.predict_proba(X_val)[:, 1]
    default_decision = (y_pred >= 0.5)
    accuracy = (y_val == default_decision).mean()
    print(y_pred[:5])

    print(regularization, accuracy, sep='\n')
    print()
```

Varying the regularization strength ($C$) alone is not sufficient to overcome the underlying class imbalance or the lack of feature scaling. While the individual probability distributions shift slightly, the changes are too minor to push any predictions across the default 0.5 decision threshold.
